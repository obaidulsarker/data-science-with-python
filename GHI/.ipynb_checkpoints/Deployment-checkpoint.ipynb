```python
import streamlit as st
import numpy as np
import cv2
import joblib
from PIL import Image
```

```python
# Load the model
model = joblib.load('models/ghi_prediction_model.pkl')
```

```python
# Image preprocessing function
def preprocess_image(image):
    # Convert the PIL image to a NumPy array
    image = np.array(image)
    # Resize the image to the required dimensions
    image = cv2.resize(image, (224, 224))
    # Normalize the image (assuming normalization was used during training)
    image = image / 255.0
    # Expand dimensions to match the model input
    image = np.expand_dims(image, axis=0)
    return image
```

```python
# Streamlit app
def main():
    st.title("GHI Prediction from Sky Images")
    st.write("Upload a sky image to predict the Global Horizontal Irradiance (GHI).")

    # File uploader
    uploaded_file = st.file_uploader("Choose an image...", type=["jpg", "jpeg", "png"])
    
    if uploaded_file is not None:
        # Display the uploaded image
        image = Image.open(uploaded_file)
        st.image(image, caption='Uploaded Image.', use_column_width=True)
        
        # Preprocess the image
        preprocessed_image = preprocess_image(image)
        
        # Make prediction
        with st.spinner('Predicting...'):
            prediction = model.predict(preprocessed_image)
            predicted_ghi = prediction[0]
        
        st.success(f'Predicted GHI: {predicted_ghi:.2f} W/m^2')

if __name__ == '__main__':
    main()
```
