```python
# Data Set

data = [110, 105, 115, 120, 110, 130, 150, 100, 105]
type(data)
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=['price'])
df
```

### Task 01: Perform Mentioned Techniques:
#### Normalization <br>
#### Standardization <br>
#### Log Transformation <br>
#### Robust Scaler <br>
#### Max Absolute Scaler

### Normalization / min max scaler

```python
min_price = df['price'].min()
max_price = df['price'].max()
```

```python
min_price
```

```python
max_price
```

```python
df['price_norm_manual'] = (df['price'] - min_price ) / (max_price - min_price)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
mmx = MinMaxScaler() #feature_range=(2,5)
```

```python
df['price_norm_sklearn'] = mmx.fit_transform(df[['price']])
```

```python
df
```

### Standardization / standard scaler

```python
mean_price = df['price'].mean()
std = df['price'].std()
```

```python
df['price_std_manual'] = (df['price'] - mean_price) / std
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
```

```python
df['price_std_sklearn'] = scaler.fit_transform(df[['price']])
```

```python
df
```

```python
std_dof = df['price'].std(ddof=0)
```

```python
std_dof
```

```python
std_dof1 = df['price'].std(ddof=1)
std_dof1
```

```python
df
```

```python
df['price_std_manual2'] = (df['price'] - mean_price) / std_dof1
```

```python
df
```

```python
df['price_std_manual3'] = (df['price'] - mean_price) / std_dof
```

```python
df
```

### Log Transformation

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
fc = FunctionTransformer(np.log1p)
df['price_fc'] = fc.fit_transform(df[['price']])
df
```

```python
df['fc_np'] = np.log1p(df[['price']])
```

```python
df
```

### Max Absolute Scaler

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
mx = MaxAbsScaler()
```

```python
df['mx_abs'] = mx.fit_transform(df[['price']])
```

```python
df
```

###  Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
RoSc=RobustScaler()
```

```python
df['rs_price'] = RoSc.fit_transform(df[['price']])
```

```python
df
```
