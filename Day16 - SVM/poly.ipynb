```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a range of x values
x = np.linspace(-3, 3, 400)

# Create a non-linear relationship for the y values
y_linear = 2 * x + np.random.normal(0, 1, size=x.shape)
y_nonlinear = np.sin(x) + np.random.normal(0, 0.2, size=x.shape)

# Create the plots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot for the linear relationship
axs[0].scatter(x, y_linear, color='blue')
axs[0].plot(x, 2 * x, color='red', linewidth=2) # plot the expected linear relationship without noise
axs[0].set_title('Linear Relationship')
axs[0].set_xlabel('Independent Variable (x)')
axs[0].set_ylabel('Dependent Variable (y)')

# Plot for the non-linear relationship
axs[1].scatter(x, y_nonlinear, color='green')
curve_x = np.linspace(-3, 3, 400)
curve_y = np.sin(curve_x)
axs[1].plot(curve_x, curve_y, color='red', linewidth=2) # plot the expected non-linear relationship without noise
axs[1].set_title('Non-Linear Relationship')
axs[1].set_xlabel('Independent Variable (x)')

plt.tight_layout()
plt.show()
```
