# Gradient Decent

```python
m = c = 0
l = .0001
n = len(x)

for i in range(500):
    ypred =( m*x) + c
    Dm = (-2/n)* sum(x*(y-ypred))
    Dc = (-2/n)* sum((y-ypred))
    
    #update
    m = m - (l*Dm)
    c = c - (l*Dc)
```
