```python
import pandas as pd

data = {
  "school_id": ["sch_1", "sch_2", "sch_3", "sch_4", "sch_5", "sch_6"],
  "state_code": ["!@sc_1", "))sc_2", "!@sc_2", "sc_2)_", "sc_1#@", "sc_10#@"],
  "subjects" : ["english maths chemistry", "english physics chemistry", "maths biology", "french", "social_studies maths literature", "english maths history"]
}

df = pd.DataFrame(data)

df
```

```python
# Function to remove special characters
def remove_special_characters(text):
    # Define a regular expression pattern to match special characters
    pattern = r'[^a-zA-Z0-9_]'
    
    # Use str.replace to remove special characters using the defined pattern
    cleaned_data = pd.Series(text).str.replace(pattern, '', regex=True)

    
    return cleaned_data
```

```python
# Apply the function to the 'Name' column
df['state_code']=remove_special_characters(df['state_code'])
```

```python
df['subjects'] = df['subjects'].str.split()
df
```

```python
# Explode all values of subject column
df_new=df.explode('subjects')
df_new
```

```python
df_new = df_new.drop('school_id', axis=1)
df_new
```

```python
df_new = df_new.reset_index(drop=True)
```

```python
result = df_new.groupby('state_code')['subjects'].aggregate('count')
result
```
